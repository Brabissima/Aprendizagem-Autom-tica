```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
```

```python
df = pd.DataFrame({'Peso': np.array([40.0, 93.5, 35.5, 30.0, 52.0, 17.0, 38.5,  8.5, 33.0,  9.5, 21.0, 79.0]) / 2.205,
                            'Altura': np.array([42.8, 63.5, 37.5, 39.5, 45.5, 38.5, 43.0, 22.5, 37.0, 23.5, 33.0, 58.0]) / 39.37,
                            'Cateter': [37, 50, 34, 36, 43, 28, 37, 20, 34, 30, 38, 47]},
                            dtype='float')
```

```python
alfa = 0.001
epocas = 10000
```

# Regressão Linear Simples 

```python
def regSimples(df):
    x = np.array(df[['Altura']])
    y = np.array(df[['Cateter']])
    y_p = np.zeros(df.shape[0])
    w = np.zeros(df.shape[1]-1)
    custos = []
    for ep in range(epocas):
        sum_erros_w0 = 0
        sum_erros_w1 = 0
        custo = 0
        erros = np.zeros(df.shape[0])
        for i in range(df.shape[0]):
            y_p[i] = w[0] + (w[1]*x[i])
            erros[i] = y[i] - y_p[i]
            custo = custo + (erros[i]**2) 
            sum_erros_w0 = sum_erros_w0 + erros[i]
            sum_erros_w1 = sum_erros_w1 + erros[i]*x[i]
        w[0] = w[0] + (alfa*(1/df.shape[0])*sum_erros_w0)
        w[1] = w[1] + (alfa*(1/df.shape[0])*sum_erros_w1)
        custos.append((1/(2*df.shape[0]))*custo)
        print("Época {}: {}".format(ep, custo))
    return pyplot.plot(custos)
```

```python
regSimples(df)
```

# Regressão Linear Múltipla

### Gradiente Descendente

```python
def gradDesc(df):
    w = np.random.uniform(size=df.shape[1])
    x = df[['Peso', 'Altura']]
    y = df['Cateter']
    x = np.array(x)
    x = (x - np.mean(x))/np.std(x)
    one = np.ones((df.shape[0],1))
    x = np.hstack((one, x))
    custos = []
    for ep in range(epocas):
        custo = 0
        grads = np.zeros(shape=w.shape)
        erros = np.zeros(df.shape[0])
        y_p = np.zeros(df.shape[0])
        for i in range(df.shape[0]):
            y_p[i] = np.dot(w.T,x[i])
            erros[i] = y[i] - y_p[i]
            custo = custo + (erros[i]**2) 
            grads = grads + erros[i]*x[i]
        w = w + (alfa * (1/len(x)) * grads)
        custos.append(((1/(2*df.shape[0]))*custo))
        print("Época {}: {}".format(ep, custo))
    return pyplot.plot(custos)
```

```python
gradDesc(df)
```

### Gradiente Descendente Estocástico

```python
def gradDescEstoc(df):
    w = np.random.uniform(size=df.shape[1])
    x = df[['Peso', 'Altura']]
    y = df['Cateter']
    x = np.array(x)
    x = (x - np.mean(x))/np.std(x)
    one = np.ones((df.shape[0],1))
    x = np.hstack((one, x))
    erro_ant = np.zeros(df.shape[0])
    custos = []
    for ep in range(epocas):
        grads = np.zeros(shape=w.shape)
        erros = np.zeros(df.shape[0])
        y_p = np.zeros(df.shape[0])
        custo = 0
        random = np.random.permutation(x.shape[0])
        x = x[random]
        y = y[random]
        for i in range(df.shape[0]):
            y_p[i] = np.dot(w,x[i])
            erros[i] = y[i] - y_p[i]
            custo = custo + (erros[i]**2)
            grads = grads + erro_ant[i]*x[i]
        w = w + (alfa * (1/len(x)) * grads)
        erro_ant = erros
        custos.append(((1/(2*df.shape[0]))*custo))
        print("Época {}: {}".format(ep, custo))
    return pyplot.plot(custos)
```

```python
gradDescEstoc(df)
```

# Regressão Linear Analítica

```python
def regLinAnalit(df):
    w = np.random.uniform(size=df.shape[1])
    x = df[['Peso', 'Altura']]
    y = df['Cateter']
    x = np.array(x)
    x = (x - np.mean(x))/np.std(x)
    one = np.ones((df.shape[0],1))
    x = np.hstack((one, x))
    y_p = np.zeros(df.shape[0])
    for i in range(df.shape[0]):
        y_p[i] = np.dot(np.transpose(w),x[i])
    w = np.linalg.inv(np.dot(np.transpose(x),x))@np.transpose(x)@y
    arr = np.zeros(df.shape[0])
    for i in range(len(arr)):
        arr[i] = y[i]
    print("y:   {}\ny_p: {}".format(arr,y_p))
    erros = np.zeros(df.shape[0])
    for i in range(df.shape[0]):
        erros[i] = y[i] - y_p[i]
    print("Erros: {}".format(erros))
```

```python
regLinAnalit(df)
```
